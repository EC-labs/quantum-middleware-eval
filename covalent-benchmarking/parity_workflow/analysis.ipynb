```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./metrics.csv')
df.head()
```

```python
# Plotting
fig, axs = plt.subplots(1, 3, figsize=(30, 8))

# Time for qubits
axs[0].plot(df.groupby('qubits')['time'].mean(), marker='o')
axs[0].set_title('Average Time for Qubits')
axs[0].set_xlabel('Qubits')
axs[0].set_ylabel('Time')
axs[0].grid(True)

# CPU for qubits
axs[1].plot(df.groupby('qubits')['cpu'].mean(), marker='o')
axs[1].set_title('Average CPU for Qubits')
axs[1].set_xlabel('Qubits')
axs[1].set_ylabel('CPU')
axs[1].grid(True)

# Memory for qubits
axs[2].plot(df.groupby('qubits')['memory'].mean(), marker='o')
axs[2].set_title('Average Memory for Qubits')
axs[2].set_xlabel('Qubits')
axs[2].set_ylabel('Memory')
axs[2].grid(True)

plt.suptitle('Effect of Qubits in Workflow', fontsize=16)

plt.tight_layout()
plt.show()
```
